## Choose Evaluation metric

To scale down the effect of the high values, RMSLE(root mean squared log error) can nullify them by using log function. 

Also, it is acceptable to have a relaxation delivery time for the customer rather than lateness. If the estimated delivery time is less than the actual arrival time, the customer reviews can be affected. Considering this situation, RMSLE can add more penalty on the model when it underestimates the time than overestimates it. Thus, the model which tends to overestimated time would preform better on the test score.

The best model should be with the minimized RMSLE.

```python
data = h5py.File('data/temp4_imputation.H5')
df_train = np.array(data['df_train'])
df_test = np.array(data['df_test'])
data.close()
```

```python
y = h5py.File('data/temp4_y.H5')
y_train = np.array(y['y_train'])
y_test = np.array(y['y_test'])
y.close()
```

```python
# machine learning pipeline
def MLpipe(X, y, ML_algo, param_grid):
    res = []
    test_scores = []
    best_models = []

    for i in range(10):
        cv = KFold(n_splits=4, shuffle=True, random_state=42 * i)

        grid = GridSearchCV(ML_algo, param_grid=param_grid, scoring='neg_mean_squared_log_error',
                            cv=cv, return_train_score=True)
        grid.fit(X, y)

        result = grid.cv_results_
        res.append(result)
        test_score = grid.score(df_test, y_test)
        test_scores.append(test_score)
        best_models.append(grid.best_params_)
    return best_models, test_scores, res
```

# Read results

```python
def read_result(best_model_, scores_, res_, pred_ = None):
    with open("result/" + best_model_+ ".txt", "rb") as fp:   # Unpickling
        best_model = pickle.load(fp)
    with open("result/" + scores_+ ".txt", "rb") as fp:   # Unpickling
        scores = pickle.load(fp)
    with open("result/" + res_+ ".txt", "rb") as fp:    # Unpickling
        res = pickle.load(fp)
    if pred_ != None:
        with open("result/" + pred_+ ".txt", "rb") as fp:    # Unpickling
            pred = pickle.load(fp)
    else:
        pred = None
    return best_model, scores, res, pred
```
