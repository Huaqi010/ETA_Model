## Preprocess the data

```python
import numpy as np
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.compose import ColumnTransformer
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import SimpleImputer
from sklearn.impute import IterativeImputer
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
import missingno as msno
from sklearn.model_selection import train_test_split 
import h5py
```

```python
df = pd.read_csv("df_filtered.csv")
df.head()
```

```python
df.dtypes
```

```python
for i in df.columns:
    print(i,len(df[i].value_counts()))
```

```python
df = df.replace(pd.NA, np.nan)
```

```python
msno.bar(df)
```

```python
msno.matrix(df)
```

```python
msno.dendrogram(df)
```

```python
df['market_id'] = df['market_id'].astype('str')
df['store_primary_category'] = df['store_primary_category'].astype('str')
df['order_protocol'] = df['order_protocol'].astype('str')
```

```python
y = df['dur']
X = df.loc[:, df.columns != 'dur']
random_state = 42

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.1 ,random_state=random_state)
print('train set:',X_train.shape, y_train.shape)
print('validation set:',X_test.shape, y_test.shape)
```

```python
random_state = 42

cat_ftrs = ['market_id','store_id_showup','store_primary_category','order_protocol','start_hour','start_month', 'start_weekday']
num_ftrs = ['total_items', 'subtotal', 'num_distinct_items', 'min_item_price','max_item_price', 'total_onshift_dashers', 'total_busy_dashers',
        'total_outstanding_orders', 'estimated_order_place_duration','estimated_store_to_consumer_driving_duration']

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant',fill_value='other')),
    ('onehot', OneHotEncoder(sparse=False,handle_unknown='ignore'))])
numeric_transformer = Pipeline(steps=[
    ('imputer2', IterativeImputer()),
    ('scaler', StandardScaler())])

preprocessor = ColumnTransformer(
    transformers=[
        ('cat', categorical_transformer, cat_ftrs),
        ('num', numeric_transformer, num_ftrs)])
```

```python
X_prep = preprocessor.fit_transform(X_train)

feature_names = list(preprocessor.named_transformers_['cat'][1].get_feature_names(cat_ftrs)) + \
                preprocessor.transformers_[1][-1]

df_train = pd.DataFrame(data=X_prep,columns=feature_names)
print(df_train.shape)

# transform the test
df_test = preprocessor.transform(X_test)
df_test = pd.DataFrame(data=df_test,columns = feature_names)
```

```python
def save_model(model, output_file):
    try:
        with open(output_file,'wb') as outfile:
            pickle.dump({
                'model':model
            },outfile)
        return True
    except:
        return False
save_model(preprocessor, 'model/preprocessor.pkl')
```

```python
h5f = h5py.File('data/temp4_imputation.H5', "w")
h5f.create_dataset('df_train', data=df_train, compression="gzip", compression_opts=9)
h5f.create_dataset('df_test', data=df_test, compression="gzip", compression_opts=9)
h5f.close()
```

```python
h5f = h5py.File('data/y.H5', "w")
h5f.create_dataset('y_train', data=y_train, compression="gzip", compression_opts=9)
h5f.create_dataset('y_test', data=y_test, compression="gzip", compression_opts=9)
h5f.close()
```

# Feature Selection by PCA

```python
import h5py
data = h5py.File('data/temp4_imputation.H5')
df_train = np.array(data['df_train'])
df_test = np.array(data['df_test'])
data.close
```

```python
y = h5py.File('data/ty.H5')
df_train = np.array(y['y_train'])
df_test = np.array(y['y_test'])
y.close
```

```python
from sklearn.decomposition import PCA
pca_30 = PCA(n_components=30).fit(df_train)
# use fit_transform to run PCA on our standardized matrix
X_train_pca30 = pca_30.transform(df_train)

save_model(pca_30, './pca_30.pkl')
# save_model(pca_100,'./pca_100.pkl')

# h5f = h5py.File('./pca', "w")
# h5f.create_dataset('pca_30', data=pca_30, compression="gzip", compression_opts=9)
# h5f.create_dataset('pca_100', data=pca_100, compression="gzip", compression_opts=9)
# h5f.close()
```

```python
with open("model/pca_30.pkl",'rb') as infile:
    loaded30 = pickle.load(infile)
pca_30 = loaded30['pca_fit']
```

```python
explained_variance=pca_30.explained_variance_ratio_
plt.subplot(211)
plt.bar(range(30), explained_variance, alpha=0.5, align='center',label='individual explained variance')
plt.ylabel('Explained variance ratio')
plt.subplot(212)
plt.plot(np.cumsum(explained_variance))
plt.ylabel('Cumulated ratio')
plt.xlabel('Principal components')
plt.tight_layout()
```

```python
feature_contribute = np.array(pca_30.components_)
weighted_contribute = (abs(feature_contribute[:5,:]).T * explained_variance[:5])
```

```python
row_sum = np.sum(weighted_contribute, axis=1)
ind = row_sum.argsort()
feature_sort = feature_contribute[:,ind]
feature_names_sort = np.array(feature_names)[ind]
```

```python
df_f = pd.DataFrame(feature_sort, columns=feature_names_sort)
df_f =df_f[df_f.columns[120:]]
```
